```python
import pandas as pd
import ast
import json

df = pd.read_csv("ebay_result.csv")
```

```python
df = df[~df["trace_id"].isnull()]
df
```

```python
expanded_rows = []
for _, row in df.iterrows():
    new_row = {
        "trace_id": row["trace_id"],
        "original_url": row["url"],
        "crawledAt": row["crawledAt"],
        "status": row["status"],
    }

    # 解析parsed欄位
    if pd.notna(row["parsed"]):
        try:
            parsed_dict = ast.literal_eval(row["parsed"])
            new_row.update(parsed_dict)
        except:
            pass

    expanded_rows.append(new_row)
```

```python
result_df = pd.DataFrame(expanded_rows)
```

```python
result_df.to_csv("ebay_result_exploded.csv", index=False)
```
